### 1. Сбор данных

```python
### Parsing data from HitBTC exchange,  
### and saving to .csv file

conn = HitBTC()

symbol = 'ETHBTC'

fy, fm, fd = '2020', '01', '01'

ty, tm, td = '2020', '04', '10'

from_date = datetime(int(fy), int(fm), int(fd), 0, 0, 0).isoformat()
to_date = datetime(int(ty), int(tm), int(td), 0, 0, 0).isoformat()

curr_date = to_date
data = []
while from_date < curr_date:
    params = {
        'sort':'DESC', 
        'limit':1000,
        'till': curr_date
    }
    
    code = 0
    while code!=200:
        response = conn.get_trades(symbol, **params)
        code = response[1]
        print(curr_date, 'Response: ', code)
        
    df = pd.DataFrame(response[0])
    curr_date = df['timestamp'].iloc[-1]
    data.append(df)
    
df = pd.concat(data)
df = df.astype({'timestamp':'datetime64[ns]'}).reset_index(drop=True)
df = df.drop(df[df['timestamp']<from_date].index)

file_name = f'{symbol}_{fy}-{fm}-{fd}_{ty}-{tm}-{td}_{df.shape[0]}.csv'
df.to_csv('data/'+file_name)

print(f'Data is saved to: {file_name}')
```

### 3. Разметка

```python
def visualize(df, data, TP, SL, size_0, size_1, point, n):

    from plotly.subplots import make_subplots
    import plotly.graph_objects as go
    import os

    case = [
        {'Timeless': 'black'},
        {'Take Profit': 'blue'},
        {'Stop Loss': 'red'},
        {'Volatility': 'yello'}
    ]
    
    df = df[n:n+size_0+size_1+1].copy()
    label = int(data[n][-1])
    result, color = case[label].popitem() 
    
    fig = make_subplots()
    
    fig.add_trace(
        go.Candlestick(
            x=df.index,
            open=df['open'], high=df['high'],
            low=df['low'], close=df['close']
        )
    )
    
    layout = {
        'title': {
            'text': "TP {}, SL {}".format(TP, SL)
        },
        'height': 500,
        'width': 900,
        'showlegend': False,
        'xaxis_rangeslider_visible': False,
        'shapes': [
            dict(
                x0=n+size_0, 
                x1=x0+size_1, 
                y0=df.loc[n+size_0]['open'], 
                y1=df.loc[n+size_0]['open']-SL/point, 
                xref='x', 
                yref='y',
                line_width=2,
                line_color='red'
            ),
            dict(
                x0=n+size_0, 
                x1=x0+size_1, 
                y0=df.loc[n+size_0]['open'], 
                y1=df.loc[n+size_0]['open']+TP/point, 
                xref='x', 
                yref='y',
                line_width=2,
                line_color='blue'
            )
        ],
        
        'annotations': [
            dict(
                x=0.02, y=0.85, 
                xref='paper', yref='paper',
                showarrow=False, xanchor='left', 
                text=result,
                font_color=color,
                font_size=20
            )
        ]
    }
    

    fig.update_layout(**layout)
    
    #fig.write_image('images/slide_{}.png'.format(n))
    fig.show()
```
